```python
#### Здесь я создал разные множества, 
#### отвечающие группам крови с разным составом антигенов,
#### и находил пересечение с объединёнными массивами антител к антигенам А и к антигенам В###
GroupA = {'A', '0'}
GroupB = {'B', '0'}
GroupAB = {'A', 'B', '0'}
Group0 = {'0'}
AntiA = {'A'}
AntiB = {'B'}

###Добавление антител в группу крови А ###
EventA = GroupA&(AntiA|AntiB)
print("""
######""")
print("Агглютинировали:",EventA)
if EventA == AntiA:
    print("Группа крови:", 'Группа A')
elif EventA == AntiB:
    print("Группа крови:",'Группа B')
elif EventA == {'AB'}:
    print("Группа крови:",'Группа AB')
else:
    print("Группа крови:",'Группа 0')
print("""######
""")

###Добавление антител в группу крови B ###
EventB = GroupB&(AntiA|AntiB)
print("""
######""")
print("Агглютинировали:",EventB)
if EventB == AntiA:
    print("Группа крови:", 'Группа A')
elif EventB == AntiB:
    print("Группа крови:",'Группа B')
elif EventB == {'AB'}:
    print("Группа крови:",'Группа AB')
else:
    print("Группа крови:",'Группа 0')
print("""######
""")

###Добавление антител в группу крови AB ###
EventAB = GroupAB&(AntiA|AntiB)
print("""
######""")
print("Агглютинировали:",EventAB)
if EventAB == AntiA:
    print("Группа крови:", 'Группа A')
elif EventAB == AntiB:
    print("Группа крови:",'Группа B')
elif EventAB == {'A', 'B'}:
    print("Группа крови:",'Группа AB')
else:
    print("Группа крови:",'Группа 0')
print("""######
""")

###Добавление антител в группу крови 0 ###
Event0 = Group0&(AntiA|AntiB)
print("""
######""")
print("Агглютинировали:",Event0)
if Event0 == AntiA:
    print("Группа крови:", 'Группа A')
elif Event0 == AntiB:
    print("Группа крови:",'Группа B')
elif Event0 == {'AB'}:
    print("Группа крови:",'Группа AB')
else:
    print("Группа крови:",'Группа 0')
print("""######
""")
```

```python
#### Здесь можно ввести антигены, присутствующие в крови, чтобы с помощью стандартного теста с аглютининами определить группу крови###


Group = set(str(input()))
AntiA = {'A'}
AntiB = {'B'}

###Добавление антител в группу крови ###
Event = Group&(AntiA|AntiB)
print("""
######""")
print("Агглютинировали:",Event)
if Event == AntiA:
    print("Группа крови:", 'Группа A')
elif Event == AntiB:
    print("Группа крови:",'Группа B')
elif Event == {'A', 'B'}:
    print("Группа крови:",'Группа AB')
else:
    print("Группа крови:",'Группа 0')
print("""######
""")
```
